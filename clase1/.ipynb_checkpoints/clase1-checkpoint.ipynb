```python
# bookcase to use
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2.5*np.pi ,20, '') # time vector
signal = np.cos(t)              # signal analog
```

```python
# Grafica de la se√±al anal√≥gica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Se√±al anal√≥gica', linestyle = '-.')
plt.stem(t, signal, 'g', label = 'Se√±al digitalizada')
plt.xlabel('Time' , fontsize = 22)
plt.ylabel('Amplitude', fontsize = 22)
plt.grid()
plt.legend()
plt.show()
```
