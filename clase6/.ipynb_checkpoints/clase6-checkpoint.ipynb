## Autocorrelación

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rect_pulse=np.ones(39) #Se genera el pulso rectangular.

Rxx=np.correlate(rect_pulse, rect_pulse, mode='full') #Se realiza la autocorrelación del pulso generado
tau=np.arange(-len(rect_pulse)+1,len(rect_pulse),1)  #Se define un vector para el eje horizontal
```

```python
plt.plot(tau, Rxx)
plt.xlabel("Tau")
plt.ylabel("Rxx")
plt.grid()
```

## Se observa que a medida que se desplaza y se traslapan las señales el valor de la correlación aumente. Pero de la misma manera disminuye, cuando las señales comienzan a ser separadas.

## El valor máximo de la correlación se dará en el momento en el que la señal que se esta moviendo se traslape totalmente. Como se puede observar el Tau = 0
