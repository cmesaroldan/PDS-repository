```python
#Importacion de las librerias necesarias para graficar y operar la libreria np, que trae consigo formas para trabajar con pi, por ejemplo
import matplotlib.pyplot as plt
import numpy as np
```

```python
f1 = 4 
f2 = 880 
fs = 45 # Frecuecnia a la cual se realizará el muestreo.

t=np.arange(0, 2.0, 1.0/fs)
x1 = np.sin(2*np.pi*f1*t)  #Señal con frecuencia igual a F1
x2 = np.sin(2*np.pi*f2*t)  #Señal con frecuencia igual a F2
plt.plot(t,x1)
plt.plot(t,x2)
plt.xlabel('Time',fontsize=18)
plt.ylabel('Amplitude',fontsize=18)
plt.legend(['Señal 1', 'Señal 2'])
plt.show()
```

### Señal 1
Cumpliendo con los requerimientos de Nyquist, logra ser bien representada.

$Fs >2fmax$

### Señal 2

Dicha señal no logra cumplir con Nyquist, pues al tener un frecuencia de 880, su frececuencia de muestreo es muy inferior a la minimamente esperada.
