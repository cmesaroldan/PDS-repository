```python
import numpy as np
import matplotlib.pyplot as plt

def region_convergencia(vector, radio=1.0):
    """
    Calcula y grafica la región de convergencia de una transformada Z a partir de un vector.
    :param vector: vector que contiene los coeficientes de la transformada Z
    :param radio: radio máximo para graficar la región de convergencia (opcional)
    """
    # Definir la variable simbólica 'z'
    z = np.complex128(np.array([1, 0]))

    # Calcular el radio de convergencia
    r = np.abs(np.roots(vector))

    # Definir el límite superior para el radio de convergencia
    r_max = min(r) if len(r) > 0 else radio

    # Definir una malla de valores de z
    N = 200
    x, y = np.meshgrid(np.linspace(-r_max, r_max, N), np.linspace(-r_max, r_max, N))
    z_values = x + 1j*y

    # Calcular la función de transferencia en la malla de valores de z
    H = np.polyval(vector, z_values) / z_values**(len(vector)-1)

    # Graficar la región de convergencia
    plt.imshow(np.abs(H) < 1, extent=(-r_max, r_max, -r_max, r_max), cmap='gray', origin='lower')
    plt.xlabel('Re(z)')
    plt.ylabel('Im(z)')
    plt.title('Región de convergencia de la transformada Z')
    plt.show()


```

```python
# Ejemplo de uso:
vector = [1, -0.5, 0.25]
region_convergencia(vector)
```
